```python
# pandasæ©Ÿèƒ½ã‚’ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
import pandas as pd
# requestsæ©Ÿèƒ½ã‚’ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
import requests
from bs4 import BeautifulSoup
```

```python
# å…ƒãƒ‡ãƒ¼ã‚¿ã®èª­ã¿è¾¼ã¿
input_df = pd.read_csv("LP_URL.csv")
```

```python
# LPãƒ†ã‚­ã‚¹ãƒˆã‚’å–å¾—ã™ã‚‹é–¢æ•°ã‚’å®šç¾©
def fetch_lp_text(request_url):
    res = requests.get(request_url, timeout=10)
    # resã®æ–‡å­—ãƒ‡ãƒ¼ã‚¿ãŒISO-8859-1ã®ã‚±ãƒ¼ã‚¹ãŒã‚ã‚‹ã®ã§ã€utf-8ã«å¤‰æ›ã—ã¦æ–‡å­—åŒ–ã‘ã‚’é˜²æ­¢
    res.encoding = "utf-8"
    soup = BeautifulSoup(res.text, "html.parser")

    # ç”»é¢éè¡¨ç¤ºã®å…¸å‹ã‚¿ã‚°ã‚’å‰Šé™¤
    for tag in soup(["script", "style", "noscript", "svg", "meta", "title", "head"]):
        tag.decompose()

    # hidden/aria-hidden/display:none/visibility:hidden ã‚’å‰Šé™¤
    for el in soup.select("[hidden], [aria-hidden='true'], [style*='display:none'], [style*='visibility:hidden']"):
        el.decompose()

    # è¡¨ç¤ºãƒ†ã‚­ã‚¹ãƒˆã®ã¿ã‚’å–å¾—ã—ã¦æ•´å½¢
    return "".join(s for s in soup.stripped_strings)
```

```python
# å‡ºåŠ›ç”¨ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã‚’ç”¨æ„ã—ã¤ã¤ã‚¹ã‚¯ãƒ¬ã‚¤ãƒ”ãƒ³ã‚°
out_df = input_df.copy()
out_df['LP_text'] = ""

for idx, row in out_df.iterrows():
    name = row['Name']
    url = row['URL']
    print(f"{idx + 1}ä»¶ç›®: {name} ã‚’ã‚¹ã‚¯ãƒ¬ã‚¤ãƒ”ãƒ³ã‚°ä¸­...")
    try:
        lp_text = fetch_lp_text(url)
    except Exception as exc:
        print(f"  å–å¾—ã«å¤±æ•—ã—ã¾ã—ãŸ: {exc}")
        lp_text = ""
    out_df.at[idx, 'LP_text'] = lp_text

out_df
```
